```python
mapping = {"未成功认证": 0, "成功认证": 1}
merged_df['Mobile Verification'] = merged_df['Mobile Verification'].map(mapping)
merged_df['Residence Verification'] = merged_df['Residence Verification'].map(mapping)
merged_df['Video Verification'] = merged_df['Video Verification'].map(mapping)
merged_df['Education Verification'] = merged_df['Education Verification'].map(mapping)
merged_df['Credit Verification'] = merged_df['Credit Verification'].map(mapping)
merged_df['Taobao Verification'] = merged_df['Taobao Verification'].map(mapping)
```

```python
merged_df['Gender'] = merged_df['Gender'].replace({"男": 0, "女": 1})
merged_df["First Bid"] = merged_df["First Bid"].replace({"是": 1, "否": 0})
```

```python
print(merged_df['Loan Type'].unique())
merged_df['Loan Type'].replace({'其他':'other','普通':'normal','电商':'E-commerce','APP闪电':'app'}, inplace=True)
print(merged_df['Loan Type'].unique())
```

```python
merged_df['Due Date'] = pd.to_datetime(merged_df['Due Date'])
merged_df['recorddate_y'] = pd.to_datetime(merged_df['recorddate_y'])
merged_df['Loan Approval Date'] = pd.to_datetime(merged_df['Loan Approval Date'])
merged_df.info()
```

The purpose of this project is to generate a model that were able to predict the repayment status, therefore columns with information regarding the investment amount shall be dropped considered they are not be haviors of borrowers.

```python
merged_df.columns
```

```python
merged_df.drop('My Investment Amount', axis=1, inplace=True)
```

```python
merged_df.head()
```

```python
#We decided to combine verifications columns by adding them up
#Recorddate is simply the date they record all these status for the borrowers, should not be influential to their behaviors, so we will remove them
merged_df.drop('recorddate_y', axis=1, inplace=True)
verification_columns = [col for col in merged_df.columns if col.endswith('Verification')]
merged_df['Verification'] = merged_df[verification_columns].sum(axis=1)
merged_df.drop(verification_columns, axis=1, inplace=True)
merged_df.columns
```

```python
#dealing with null values
merged_df.isna().sum()
```

```python
#Last Payment Date is null, indicating that they did were overdue for the last installment
#Next Due date is null, probably due to already paid or haven't paid for last installment
#repayment status is our targer column, we are predicting it based on the past behavior, so the next due columns are unnecessary
nextduecols = [col for col in merged_df.columns if col.startswith('Next Due')]
merged_df.drop(nextduecols, axis=1, inplace=True)
```

```python
rows = merged_df[merged_df.isnull().any(axis=1)]
rows['Bid Status'].unique()
```

```python
#Given that bid status aligns with our assumption on the null rows, we can remove the last paid columns
lastpaidcols = [col for col in merged_df.columns if col.startswith('Last ')]
merged_df.drop(lastpaidcols, axis=1, inplace=True)
```

```python
merged_df.isna().sum()
```

1. Separate Loan Approval Date based on quarter
2. Since Principal Paid and Principal wait for paid, Interest Paid and Interest wait for paid, indicating the same information, we can remove them, same for principal due and remaining principal due, interest due and remaining interest.
3. We will keep remaining principal due and remaining interest, as principal due and interest due can be calculated based on interest rate and loan term
4. Due date and repayment date can be combined to overdue to indicate if overdue in the current term
5. Last, we will use graphical analysis to determine the grouping of age
6. further graphical analysis will be conducted to generate a portrait of users
6. Before modelling, we will use decision tree to find weightage of different factors and remove the less important one

```python
merged_df['Loan Approval Date'] = merged_df['Loan Approval Date'].dt.to_period('Q').astype(str).str[-2:]
merged_df.head()
```

```python
merged_df['Principal Paid PCT'] = merged_df['Principal Paid']/(merged_df['Principal Paid']+merged_df['Principal wait for paid'] *100)
merged_df['Interest Paid PCT'] = merged_df['Interest Paid']/(merged_df['Interest Paid'] + merged_df['Interest wait for paid'] *100)
merged_df.drop(['Principal Paid', 'Principal wait for paid', 'Interest Paid', 'Interest wait for paid'], axis=1, inplace=True)
merged_df.head()
```

```python
merged_df.drop(['Principal Due', 'Interest Due'], axis=1, inplace=True)
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df['Repayment Date'].unique
```

```python
merged_df['Repayment Date'] = merged_df['Repayment Date'].replace(r'\N', None)
```

```python
merged_df['Repayment Date'] = pd.to_datetime(merged_df['Repayment Date'])
merged_df['OverDue'] = (merged_df['Repayment Date'].isna()) | (merged_df['Repayment Date'] > merged_df['Due Date'])
merged_df.drop(['Repayment Date', 'Due Date'], axis=1, inplace=True)
```

```python
merged_df['OverDue'] = merged_df['OverDue'].astype("int")
```

```python
merged_df.columns
```

```python
#now access the target column: Repayment Status
# 0-‘未还款’，1-‘已正常还款’，2-‘已逾期还款’，3-‘已提前还清该标全部欠款’，4-‘已部分还款’
# Now combining 0 and 2 as "1", 1 and 3 as "0"
merged_df['Repayment Status'].replace({0:1,2:1,1:0,3:0},inplace=True)
merged_df.head()
```

```python
merged_df['Bid Status'].unique()
merged_df['Bid Status'].replace({'已还清': 0, '正常还款中':0 , '逾期中':1}, inplace=True)
```
